````python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return '''
    <!DOCTYPE html>
    <html>
    <head>
        <title>Kutty - Small Milestone</title>
        <style>
            body { font-family: Arial, sans-serif; background: #f0f0f0; }
            h1 { color: #333; }
        </style>
    </head>
    <body>
        <h1>Hello, Kutty!</h1>
        <p>This is a simple Flask app serving HTML directly from Python code.</p>
        <button onclick="alert('Hello from JS!')">Click Me</button>
    </body>
    </html>
    '''

if __name__ == '__main__':
    app.run(debug=True, port=1234)
````
